### Single Source

#### Nature of the Path

```python
#Importing useful packges for generating the results
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Defining problem's parameters 
dt=0.001
u=0.02
#position of source
x0=0
y0=0
#Number of iterations
N=40000
#Define empty lists for storing generated data points in iterations, and also defining initial postions for inclusion.
y=np.zeros(N)
x=np.zeros(N)
x[0]=2
y[0]=1
y1=np.zeros(N)
x1=np.zeros(N)
x1[0]=0
y1[0]=1
y2=np.zeros(N)
x2=np.zeros(N)
x2[0]=1
y2[0]=0
y3=np.zeros(N)
x3=np.zeros(N)
x3[0]=-2
y3[0]=-1
t=np.zeros(N)
t[0]=dt
```

#### We define the length scale For the inclusion 

```python
ld = np.sqrt((x[0]-x0)**2 + (y[0]-y0)**2)
ld1 = np.sqrt((x1[0]-x0)**2 + (y1[0]-y0)**2)
ld2= np.sqrt((x2[0]-x0)**2 + (y2[0]-y0)**2)
ld3 = np.sqrt((x3[0]-x0)**2 + (y3[0]-y0)**2)
```

#### We generate the trajectories 


```python
for i in range(1,N):
    t[i]=i*dt
    x[i]=x[i-1]+((u/ld)*(x[i-1]-x0)*np.exp((-(x[i-1]-x0)**2 -(y[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y[i]=y[i-1]+((u/ld)*(y[i-1]-y0)*np.exp((-(x[i-1]-x0)**2 -(y[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
t=np.zeros(N)
t[0]=dt    
for i in range(1,N):
    t[i]=i*dt
    x1[i]=x1[i-1]+((u/ld1)*(x1[i-1]-x0)*np.exp((-(x1[i-1]-x0)**2 -(y1[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y1[i]=y1[i-1]+((u/ld1)*(y1[i-1]-y0)*np.exp((-(x1[i-1]-x0)**2 -(y1[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
t=np.zeros(N)
t[0]=dt
for i in range(1,N):
    t[i]=i*dt
    x2[i]=x2[i-1]+((u/ld2)*(x2[i-1]-x0)*np.exp((-(x2[i-1]-x0)**2 -(y2[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y2[i]=y2[i-1]+((u/ld2)*(y2[i-1]-y0)*np.exp((-(x2[i-1]-x0)**2 -(y2[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
t=np.zeros(N)
t[0]=dt
for i in range(1,N):
    t[i]=i*dt
    x3[i]=x3[i-1]+((u/ld3)*(x3[i-1]-x0)*np.exp((-(x3[i-1]-x0)**2 -(y3[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y3[i]=y3[i-1]+((u/ld3)*(y3[i-1]-y0)*np.exp((-(x3[i-1]-x0)**2 -(y3[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
```

#### Plotting Trajectory and displacement components for the inclusion when initially at (2,1)

```python
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('x')
plt.title('Displacement Along x-axis')
#plt.savefig('Tx21.pdf')
```

```python
plt.plot(t,y1)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Displacement Along y-axis')
#plt.savefig('Ty21.pdf')
```

```python
#plotting trajectory
plt.plot(x,y)
plt.title('Trajectory Of Inclusion')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('T21.pdf')
```

#### Plotting trajectories of inclusion when from 4 different initial positions (plots made on same axes)

```python
plt.plot(x,y,label='From (2,1)')
plt.plot(x1,y1,label='From (0,1)')
plt.plot(x2,y2,label='From (1,0)')
plt.plot(x3,y3,label='From (-2,-1)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trajectories Of Inclusion')
plt.savefig('T4.pdf')
```

#### We check dependence of displacement on initial distance of inclusion from source

```python
y=np.zeros(N)                     # two lists to store the data as it is generated by the codes
x=np.zeros(N)
x0=0                              # position of source: (x0,y0)
y0=0
x[0]=0                            # Initial position of inlcusion : (x[0], y[0])  
y[0]=1
y1=np.zeros(N)                    # two morelists to store data for the numerical solution for the second case.
x1=np.zeros(N)
x1[0]=0
y1[0]=1
ld = np.sqrt((x[0]-x0)**2 + (y[0]-y0)**2)      # length scale for the problem
t=np.zeros(N)
t[0]=dt
for i in range(1,N):          # for loops generating numerical solution to equations of motion for the two cases
    t[i]=i*dt
    x[i]=x[i-1]+((u/ld)*(x[i-1]-x0)*np.exp((-(x[i-1]-x0)**2 -(y[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y[i]=y[i-1]+((u/ld)*(y[i-1]-y0)*np.exp((-(x[i-1]-x0)**2 -(y[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2

x0=0           # source position for the second case          
y0=-1
ld1 = np.sqrt((x1[0]-x0)**2 + (y1[0]-y0)**2)
t=np.zeros(N)
t[0]=dt
for i in range(1,N):
    t[i]=i*dt
    x1[i]=x1[i-1]+((u/ld1)*(x1[i-1]-x0)*np.exp((-(x1[i-1]-x0)**2 -(y1[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
    y1[i]=y1[i-1]+((u/ld1)*(y1[i-1]-y0)*np.exp((-(x1[i-1]-x0)**2 -(y1[i-1]-y0)**2)/(4*t[i-1])))/(4*t[i-1])**2
```

```python
plt.plot(t,y,label='Source at (0,0)')
plt.plot(t,y1,label='Source at (0,-1)')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Displacement Along y-axis')
plt.savefig('TTy.pdf')
```

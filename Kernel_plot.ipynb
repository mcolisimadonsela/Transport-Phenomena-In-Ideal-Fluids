```python
import numpy as np
import matplotlib.pyplot as plt
```

#### With Boundary

```python
x = np.linspace(0,10,200)
D= 1
x1=2.0
t1=0.2
t2=0.3
t3=0.5
t4=1
t5=10
K1 = np.zeros(200)
K2 = np.zeros(200)
K3 = np.zeros(200)
K4 = np.zeros(200)
K5 = np.zeros(200)

for i in range(200):
    a=(x[i]-x1)**2
    b=(x[i]+x1)**2
    K1[i] = (1/(np.sqrt(4*D*np.pi*t1)))*(np.exp(-a/(4*D*t1)) + np.exp(-(b/(4*D*t1))))
    
for i in range(200):
    a=(x[i]-x1)**2
    b=(x[i]+x1)**2
    K2[i] = 1/(np.sqrt(4*D*np.pi*t2))*(np.exp(-a/(4*D*t2)) + np.exp(-b/(4*D*t2)))    

for i in range(200):
    a=(x[i]-x1)**2
    b=(x[i]+x1)**2
    K3[i] = 1/(np.sqrt(4*D*np.pi*t3))*(np.exp(-a/(4*D*t3)) + np.exp(-b/(4*D*t3)))    

for i in range(200):
    a=(x[i]-x1)**2
    b=(x[i]+x1)**2
    K4[i] = (1/(np.sqrt(4*D*np.pi*t4)))*(np.exp(-a/(4*D*t4)) + np.exp(-b/(4*D*t4)))    
    
for i in range(200):
    a=(x[i]-x1)**2
    b=(x[i]+x1)**2
    K5[i] = 1/(np.sqrt(4*D*np.pi*t5))*(np.exp(-a/(4*D*t5)) + np.exp(-b/(4*D*t5)))        
```

```python
plt.plot(x,K1,label="t=0.2")
plt.plot(x,K2,label='t=0.3')
plt.plot(x,K3,label="t=0.5")
plt.plot(x,K4,label="t=1")
plt.plot(x,K5,label="t=10")
plt.legend()
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Kernel For Different Values of t With Boundary')
#plt.show()
plt.savefig('Kernel1.png')
```
